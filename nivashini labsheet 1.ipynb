```python
# STEP 1: OPEN the file, count_2w.txt
with open('count_2w.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()

# STEP 2: build goog2w_list
goog2w_list = []
for line in lines:
    words, count = line.split('\t')
    w1, w2 = words.split()
    goog2w_list.append(((w1, w2), int(count)))

# STEP 3: build goog2w_fd
import nltk
goog2w_fd = nltk.FreqDist([x[0] for x in goog2w_list])

# STEP 4: explore
# What are the top-10 bigrams?
print("Top-10 bigrams:")
print(goog2w_fd.most_common(10))

# What are the top so-initial bigrams?
so_bigrams = [x for x in goog2w_fd if x[0] == 'so']
so_bigrams_fd = nltk.FreqDist(so_bigrams)
print("\nTop so-initial bigrams:")
print(so_bigrams_fd.most_common())

# Back to those bigrams necessary for computing the probability of the sentence 'She was not afraid.'. Are they all found in this data?
bigrams = [('she', 'was'), ('was', 'not'), ('not', 'afraid')]
for bigram in bigrams:
    if bigram not in goog2w_fd:
        print(f"\nBigram {bigram} not found in data.")

# Find a bigram that you think should be represented and it is.
print("\nBigram that should be represented and it is:")
print(goog2w_fd['(the, end)'])

# Find a bigram that you think should be represented but is not.
print("\nBigram that should be represented but is not:")
print(goog2w_fd['(the, beginning)'])

# STEP 5: pickle the data
import pickle
with open('goog2w_list.pkl', 'wb') as f:
    pickle.dump(goog2w_list, f)

with open('goog2w_fd.pkl', 'wb') as f:
    pickle.dump(goog2w_fd, f)
```
