# Question 2: Length

How many Emma word types are 15 characters or longer? Exclude hyphenated words.



```python

length = [word for word in etypes if len(word) >= 15 and '-' not in word]
print(length)
```

# Question 3: Average word length

What's the average length of all Emma word types?



```python

avg_word_length = sum(len(word) for word in etypes) / len(etypes)

print(avg_word_length)
```

# Question 4: Word frequency

How many Emma word types have a frequency count of 200 or more? How many word types appear only once?



```python

freq_200_or_more = len([word for word in etypes if efreq[word] >= 200])
single_occurrence_words = len([word for word in etypes if efreq[word] == 1])
print(freq_200_or_more )
print(single_occurrence_words)
```
