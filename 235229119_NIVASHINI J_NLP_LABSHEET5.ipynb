```python
import nltk
```

# FILE 1: austen-emma.txt

# A. Open the text file, read in the string content

```python
with open('gift-of-magi.txt', 'r') as file:
    content = file.read()
```

# B. Build a list of individual sentences

```python
import nltk
nltk.download('punkt')
  
```

```python
sentences = nltk.sent_tokenize(content)
sentences
```

# C. Print out how many sentences there are

```python
print(f"There are {len(sentences)} sentences in the corpus.")
```

# D. Build a flat tokenized word list and the type list

```python
words = nltk.word_tokenize(content)
types = set(words)
types, words




```

# E. Print the token and the type counts of this corpus

```python
print(f"There are {len(words)} tokens and {len(types)} types in the corpus.")
```

# F. Build a frequency count dictionary of words

```python
freq_dict = nltk.FreqDist(words)
freq_dict
```

# G. Print the top 50 word types and their counts

```python
print("The top 50 word types and their counts are:")
for word, count in freq_dict.most_common(50):
    print(f"{word}: {count}")
```

```python
# For example, let's find the average word length in the corpus
average_word_length = sum(len(word) for word in words) / len(words)
print(f"Average word length: {average_word_length:.2f} characters")
```
